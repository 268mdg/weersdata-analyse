```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/wereldbevolking.csv')
```

```python
df.head()
```

```python
jaren = ['1950', '1960', '1970', '1980', '1990', '2000', '2010', '2020', '2023']

China = df[df['land'] == 'China'][jaren].values[0] / 1_000_000
India = df[df['land'] == 'India'][jaren].values[0]
Verenigde_Staten = df[df['land'] == 'Verenigde Staten'][jaren].values[0]
Nederland = df[df['land'] == 'Nederland'][jaren].values[0]
Japen = df[df['land'] == 'Japan'][jaren].values[0]
```

```python
plt.plot(jaren, China, label='China')
plt.plot(jaren, India, label='India')
plt.plot(jaren, Verenigde_Staten, label='Verenigde Staten')
plt.plot(jaren, Nederland, label='Nederland')
plt.plot(jaren, Japen, label='Japan')

plt.xlabel('Jaren')
plt.ylabel('Bevolking in miljoenen')

plt.legend()

plt.show
```
