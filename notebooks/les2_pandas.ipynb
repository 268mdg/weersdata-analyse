```python
import pandas as pd

print("Pandas versie:", pd.__version__)
```

```python
data = {
    "datum": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "stad": ["De Bilt", "De Bilt", "De Bilt", "De Bilt", "De Bilt"],
    "temperatuur": [3.2, 4.1, 2.8, 1.5, 5.0],
    "neerslag_mm": [1.4, 0.0, 3.2, 0.5, 0.0]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Hoeveel rijen en kolommen?
print("Vorm:", df.shape)

# Wat voor soort data zit erin?
print(df.dtypes)

# Basisstatistieken
print(df.describe())
```

```python
# Dagen waarop het geregend heeft
regen = df[df["neerslag_mm"] > 0]
print("Regendagen:")
print(regen)
```

```python
# Gemiddelde temperatuur
gemiddeld = df["temperatuur"].mean()
print(f"Gemiddelde temperatuur: {gemiddeld:.1f} °C")

# Koudste dag
koudste = df[df["temperatuur"] == df["temperatuur"].min()]
print("Koudste dag:")
print(koudste)
```
