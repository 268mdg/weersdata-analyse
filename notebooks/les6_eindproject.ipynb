```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/weersdata.csv")
```

```python
gemiddelde_temperatuur = np.mean(df['temperatuur'])
maximum_temperatuur = np.max(df['temperatuur'])
minimum_temperatuur = np.min(df['temperatuur'])
totale_neerslag = np.sum(df['neerslag_mm'])
gemiddelde_windsnelheid = np.mean(df['windsnelheid'])

print(f"Gemiddelde temperatuur: {gemiddelde_temperatuur:.1f} °C")
print(f"Maximum temperatuur: {maximum_temperatuur:.1f} °C")
print(f"Minimum temperatuur: {minimum_temperatuur:.1f} °C")
print(f"Totale neerslag: {totale_neerslag:.1f} mm")
print(f"Gemiddelde windsnelheid: {gemiddelde_windsnelheid:.1f} km/h")
```

```python
warme_dagen = df[df['temperatuur'] > 5]
neerslag_dagen = df[df["neerslag_mm"] > 2]

print(f"Dagen warmer dan 5 °C: {warme_dagen}")
print(f"Dagen met meer dan 2 mm neerslag: {neerslag_dagen}")
```

```python
plt.title("Temperatuur over de tijd")
plt.xlabel("Datum")
plt.ylabel("Temperatuur (°C)")

plt.xticks(rotation=45)

plt.plot(df['datum'], df['temperatuur'])
```

```python
plt.title("Neerslag over de tijd")
plt.xlabel("Datum")
plt.ylabel("Neerslag (mm)")

plt.xticks(rotation=45)

plt.bar(df['datum'], df['neerslag_mm'])
```

```python
plt.title("Windsnelheid vs Temperatuur")
plt.xlabel("Windsnelheid (km/h)")
plt.ylabel("Temperatuur (°C)")

plt.scatter(df['windsnelheid'], df['temperatuur'])
```
