```python
import pandas as pd

df = pd.read_csv("../data/weersdata.csv")
```

```python
df
```

```python
import numpy as np
```

```python
gemiddelde = np.mean(df['temperatuur'])
print(f"Gemiddelde: {gemiddelde:.1f} °C")

max_t = np.max(df['temperatuur'])
print(f"Maximum: {max_t} °C")

min_t = np.min(df['temperatuur'])
print(f"Minimum: {min_t} °C")

standaard_afwijking = np.std(df['temperatuur'])
print(f"Standaard afwijking: {standaard_afwijking:.1f} °C")
```
