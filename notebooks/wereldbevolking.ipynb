# Wereldbevolkging data project

importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

datafram

```python
df = pd.read_csv('../data/wereldbevolking.csv')
```

```python
df.head()
```

data instellen

```python
jaren = ['1950', '1960', '1970', '1980', '1990', '2000', '2010', '2020', '2023']

China = df[df['land'] == 'China'][jaren].values[0] / 1_000_000
India = df[df['land'] == 'India'][jaren].values[0] /1_000_000
Verenigde_Staten = df[df['land'] == 'Verenigde Staten'][jaren].values[0] /1_000_000
Nederland = df[df['land'] == 'Nederland'][jaren].values[0] /1_000_000
Japen = df[df['land'] == 'Japan'][jaren].values[0] /1_000_000
```

totale groei per land berekenen

```python
totale_groei_China = (China[-1] - China[0]) / China[0] * 100
totale_groei_India = (India[-1] - India[0]) / India[0] * 100
totale_groei_Verenigde_Staten = (Verenigde_Staten[-1] - Verenigde_Staten[0]) / Verenigde_Staten[0] * 100
totale_groei_Nederland = (Nederland[-1] - Nederland[0]) / Nederland[0] * 100
totale_groei_Japan = (Japen[-1] - Japen[0]) / Japen[0] * 100

df['totale_groei'] = [totale_groei_China, totale_groei_India, totale_groei_Verenigde_Staten, totale_groei_Nederland, totale_groei_Japan]
```

## Grafieken

```python
plt.plot(jaren, China, label='China')
plt.plot(jaren, India, label='India')
plt.plot(jaren, Verenigde_Staten, label='Verenigde Staten')
plt.plot(jaren, Nederland, label='Nederland')
plt.plot(jaren, Japen, label='Japan')

plt.title('Bevolking van China, India, Verenigde Staten, Nederland en Japan van 1950 tot 2023')
plt.xlabel('Jaren')
plt.ylabel('Bevolking in miljoenen')

plt.legend()

plt.show
```

```python
plt.bar(df['land'], (df['2023'] / 1_000_000), color='green')

plt.title('Bevolking van China, India, Verenigde Staten, Nederland en Japan in 2023')
plt.xlabel('Land')
plt.ylabel('Bevolking in miljoenen')
plt.xticks(rotation=45)

plt.show
```

```python
plt.bar(df['land'], df['totale_groei'], color='orange')

plt.title('Bevolkingsgroei van China, India, Verenigde Staten, Nederland en Japan van 1950 tot 2023')
plt.xlabel('Land')
plt.ylabel('Groei in procenten')
plt.xticks(rotation=45)

plt.show
```
