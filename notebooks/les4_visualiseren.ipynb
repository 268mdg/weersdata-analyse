```python
import pandas as pd

df = pd.read_csv("../data/weersdata.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Temperatuur over de Datums")
plt.xlabel("Datum")
plt.ylabel("Temperatuur (°C)")

plt.xticks(rotation=45)

plt.plot(df["datum"], df["temperatuur"])
```

```python
plt.title("Neerslag over de Datums")
plt.xlabel("Datum")
plt.ylabel("Neerslag (mm)")

plt.xticks(rotation=45)

plt.bar(df["datum"], df["neerslag_mm"])
```

```python
plt.title("Windsnelnheid vs Temperatuur")
plt.xlabel("Windsnelheid (km/h)")
plt.ylabel("Temperatuur (°C)")

plt.scatter(df["windsnelheid"], df["temperatuur"])
```
